```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy
from nltk.corpus import wordnet

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
nltk.download('wordnet')

# User Input
print("Enter the News Article:")
news = input()
print("\n" + "=" * 50)
print("SMART NEWS ANALYSIS SYSTEM")
print("=" * 50)

# Tokenization & POS Tagging
tokens = word_tokenize(news)
tagged = pos_tag(tokens)

# 1. Parsing
print("\n1. Parsing")
subject = ""
verb = ""
obj = ""
# Find Subject (Proper Noun)
for word, tag in tagged:
    if tag == "NNP":
        subject = word
        break
# Find Verb
for word, tag in tagged:
    if tag.startswith("VB"):
        verb = word
        break
# Find Object 
for i in range(len(tagged) - 1):
    if tagged[i][1] == "JJ" and tagged[i + 1][1].startswith("NN"):
        obj = tagged[i][0] + " " + tagged[i + 1][0]
        break
print("Subject :", subject)
print("Verb    :", verb)
print("Object  :", obj)

# 2. Named Entity Recognition
print("\n2. Named Entities")
tree = ne_chunk(tagged)
found_entity = False
for chunk in tree:
    if isinstance(chunk, Tree):
        found_entity = True
        entity = " ".join(c[0] for c in chunk)
        print(chunk.label(), ":", entity)
if not found_entity:
    print("No named entities found.")

# 3. News Classification
print("\n3. News Classification")
train_data = [
    ({"ai": True, "technology": True, "microsoft": True}, "Technology"),
    ({"cricket": True, "match": True, "player": True}, "Sports"),
    ({"government": True, "minister": True, "election": True}, "Politics"),
    ({"business": True, "market": True, "stock": True}, "Business")
]
classifier = NaiveBayesClassifier.train(train_data)
features = {
    "ai": "ai" in news.lower(),
    "technology": "technology" in news.lower(),
    "microsoft": "microsoft" in news.lower(),
    "business": "business" in news.lower(),
    "market": "market" in news.lower(),
    "stock": "stock" in news.lower(),
    "cricket": "cricket" in news.lower(),
    "match": "match" in news.lower(),
    "player": "player" in news.lower(),
    "government": "government" in news.lower(),
    "minister": "minister" in news.lower(),
    "election": "election" in news.lower()
}
category = classifier.classify(features)
print("Category :", category)

# 4. Model Evaluation
print("\n4. Model Evaluation")
test_set = [
    ({"ai": True, "technology": True, "microsoft": True}, "Technology"),
    ({"business": True, "market": True, "stock": True}, "Business")
]
acc = accuracy(classifier, test_set)
correct = 0
for feature, label in test_set:
    prediction = classifier.classify(feature)
    if prediction == label:
        correct += 1
precision = correct / len(test_set)
print("Accuracy :", round(acc * 100, 2), "%")
print("Precision:", round(precision * 100, 2), "%")

# 5. Word Frequency & Probability Distribution
print("\n5. Word Frequency")
words = [w.lower() for w in tokens if w.isalpha()]
freq = FreqDist(words)
for word, count in freq.items():
    print(word, ":", count)
print("\nProbability Distribution")
total = len(words)
for word in freq:
    print(f"P({word}) = {freq[word]}/{total} = {freq[word]/total:.2f}")

# 6. Semantic Analysis
print("\n6. Semantic Analysis")
important_words = ["technology", "business", "platform", "ai"]
for word in important_words:
    print("\nWord :", word)
    if word.lower() == "ai":
        print("Meaning : Artificial Intelligence")
    else:
        synsets = wordnet.synsets(word)
        if synsets:
            print("Meaning :", synsets[0].definition())
        else:
            print("Meaning not found.")

# 7. Bigrams
print("\n7. Bigrams")
for bg in bigrams(words):
    print(bg)

# 8. Trigrams
print("\n8. Trigrams")
for tg in trigrams(words):
    print(tg)
```
